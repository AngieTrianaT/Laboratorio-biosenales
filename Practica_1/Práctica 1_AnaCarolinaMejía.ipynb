### a) Creación del par de vectores (a y b)

```python
a = np.array([3.1,1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
```

### b) Producto escalar de a,b.

Para poder realizar la multiplicación escalar entre un par de arrays de manera que el resultado esté definido, el número de columnas del primero debe coincidir con el número de filas del segundo.  

```python
prod = np.dot(a,b)
print(prod)
```

## c) Producto punto a punto

Para que sea posible la multiplicación punto por punto, es necesario que los arrays coincidan en parámetros como la dimensión y el tamaño. (ndim, size)

```python
mult = a * b
print(mult)
```

## f) Función de los comandos

### ones: 
El comando numpy.ones, crea una matriz cuyos elementos tendrán el valor 1.

### round:
Redondea uniformemente los números contenidos en la matriz.

### ceil:
Opera la matriz de manera que calcula el entero más cercano hacia arriba de cada uno de los elementos que la conforman.

### floor: 
Opera la matriz de manera que calcula el entero más cercano hacia abajo de cada uno de los elementos que la conforman.

### Ejemplificación

```python
#numpy.ones
matriz = np.ones((2,13), dtype='float') #Se crea una matriz de unos, con dos filas y 13 columnas. 
print(matriz)
```

```python
array = np.random.rand(2,3) #Se crea una matriz de valores aleatorios, con dos filas y 3 columnas.
print(f'''Matriz original 

{array}''')
```

```python
#numpy.round 
print(f'''Matriz - comando np.round 

{np.round(array)}''')
```

```python
#numpy.ceil
print(f'''Matriz - comando np.ceil 

{np.ceil(array)}''')
```

```python
#numpy.floor
print(f'''Matriz - comando np.floor 

{np.floor(array)}''')
```

## j) Construcción de la función y[n]=sin(π*0.12n) en el intervalo 0 <= n <= 100.

```python
n = np.arange(0,100) #Se crea la matriz que se encuentre en el rango de 0 a 100. 
yn = np.sin(np.pi*0.12*n)
```

```python
yn
```

## k) Construcción función senoidal y2[n]=cos⁡(2π*0.03n) 

```python
y2=np.cos(2*np.pi*0.03*n)
```

## l) Suma y producto de las dos funciones

```python
s=yn+y2 #suma 
t=yn*y2 #producto
```

## m) Gráficas de las señales iniciales

```python
#Forma 1 
fig_1, ax= plt.subplots(2)
ax[0].plot(yn, color='#8A2BE2')
ax[0].set_title('y[n] = sin⁡(π*0.12n)')
ax[0].set_xlim(0,100)
ax[0].set_ylim(-1,1)
ax[0].set_xlabel('Intervalo')
ax[0].set_ylabel('Amplitud')
ax[1].set_title('y2[n] = cos⁡(2π*0.03n)')
ax[1].set_ylim(-1,1)
ax[1].set_xlim(0,100)
ax[1].set_xlabel('Intervalo')
ax[1].set_ylabel('Amplitud')
ax[1].plot(y2, color='#00FFFF')
ax[0].grid()
ax[1].grid()
plt.tight_layout()
```

```python
#Forma 2
fig_1, ax1= plt.subplots(1)
plt.tight_layout()
ax1.set_title('Funciones')
ax1.set_xlabel('Intervalo')
ax1.set_ylabel('Amplitud')
ax1.plot(yn, color='#8A2BE2')
ax1.plot(y2, color='#00FFFF')
ax1.set_xlim(0,100)
ax1.set_ylim(-1.05,1.3)
ax1.legend(['y[n] = sin⁡(π*0.12n)','y2[n] = cos⁡(2π*0.03n)'], loc='upper right')
```

## n) Gráficas de la suma y producto entre funciones y[n] y y2[n]

```python
#Forma 1 
fig_1, ax= plt.subplots(2)
ax[0].set_title('Suma y[n] + y2[n]')
ax[0].set_xlabel('Intervalo')
ax[0].set_ylabel('Amplitud')
ax[1].set_title('Producto y[n] * y2[n]')
ax[1].set_xlabel('Intervalo')
ax[1].set_ylabel('Amplitud')
ax[0].plot(s, color='#FFD700')
ax[1].plot(t, color='#FF7F50')
ax[0].grid()
ax[1].grid()
plt.tight_layout()
```

```python
#Forma 2
fig_1, ax1= plt.subplots(1)
plt.tight_layout()
plt.grid()
ax1.set_title('Producto y suma de las funciones')
ax1.plot(s, color='#FFD700')
ax1.plot(t, color='#FF7F50')
ax1.set_xlim(0,99)
ax1.set_xlabel('Intervalo')
ax1.set_ylabel('Amplitud')
ax1.legend(['y[n] + y2[n]', 'y[n] * y2[n]'])
```

```python
import pandas as pd
```

```python
def notas(n): #n corresponde al número de notas que desea ingresar.
    dict = {}
    for i in range(n):
        cedula = input('Ingrese la cédula del estudiante:')
        nota_final = float(input('Nota final del estudiante:'))
        if cedula not in dict:
            dict[cedula] = nota_final
        else:
            print('Ya existe esa clave en el sistema')
            continue
            
    notas_alumnos = pd.Series(dict)
    serie = {'Nota mínima':notas_alumnos.min(), 'Nota máxima':notas_alumnos.max(), 'Promedio':notas_alumnos.mean(), 'Desviación típica': notas_alumnos.std()}
    resultado = pd.Series(serie)
    return resultado
```

```python
notas(2) #2 = número de estudiantes a los que se les registrará la nota
```
