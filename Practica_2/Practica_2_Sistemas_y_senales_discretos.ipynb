1. Función escalón unitario


```python
def unseq(no,n1,n2):
  n=np.arange(n1,n2+1)
  for i in n:
    if i>=no:
      n[i]=1
    elif i<no:
      n[i]=n[i]*0
  return n
```

```python
resul1=unseq(5,0,7)
print(resul1)
plt.stem(resul1)
plt.xlabel("n")
plt.ylabel("u(n)")
plt.title("Función Escalón Unitario")
plt.show()
```

2. Función rampa

```python
def ramseq(no,n1,n2):
  n=np.arange(n1,n2+1)
  for i in n:
    if i<no:
      n[i]=n[i]*0
    elif i>=no:
      n[i]=n[i]-no
  return n
```

```python
resul2=ramseq(5,0,10)
print(resul2)
plt.stem(resul2)
plt.xlabel("n")
plt.ylabel("r(n)")
plt.title("Función Rampa")
plt.show()
```

3. Impulso

```python
def impseq(no,n1,n2,a,n):

  for i in n:
    if i==no:
      n[i]=1*a
    else:
      n[i]=n[i]*0
  return n
```

```python
n=np.arange(0,10)
resul3=impseq(2,0,10,3,n)
result4=impseq(0,0,10,5,n)
result5=impseq(1,0,10,3,n)
plt.stem(result5,result4,resul3)
plt.xlabel("n")
plt.ylabel("r(n)")
plt.title("Función Impulso")
plt.show()
```
